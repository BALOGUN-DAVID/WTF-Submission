# Numpy & Pandas In Built Methods

### Pandas

```python
import numpy as np
import pandas as pd
```

```python
dir(pd)
```

1. pd.array: Used to create an array.

2. pd.bdate_range: Used to return a fixed frequency DatetimeIndex, with business day as the default frequency.

3. pd.concat: used to concatenate pandas objects along a particular axis with optional set logic along the other axes.

4. pd.cut: Used to bin values into discrete intervals. It is also used when you need to segment and sort data values into bins.

5. pd.date_range: Used to return a fixed frequency DatetimeIndex.

6. pd.describe_option: Prints the description for one or more registered options.

7. pd.eval: Evaluate a Python expression as a string using various backends.

8. pd.factorize: Encode the object as an enumerated type or categorical variable.

9. pd.get_dummies: Convert categorical variable into dummy/indicator variables.

10. pd.get_option: Retrieves the value of the specified option.

11. pd.interval_range: Return a fixed frequency IntervalIndex.

12. pd.isna: Detect missing values for an array-like object.

13. pd.isnull: This function takes a scalar or array-like object and indicates whether values are missing

14. pd.lreshape: Reshape wide-format data to long. Generalized inverse of DataFrame.pivot.

15. pd.unique: Returns unique values in order of appearance. This does NOT sort.

16. pd.merge_ordered: Perform merge with optional filling/interpolation.

17. pd.notna: This function takes a scalar or array-like object and indicates whether there are non-missing values.

18. pd.to_datetime: Convert argument to datetime.

19. pd.value_counts: Compute a histogram of the counts of non-null values.

20. pd.pivot: Return reshaped DataFrame organized by given index / column values.

### Numpy

```python
dir(np)
```

1. np.abs: Calculate the absolute value element-wise.

2. np.add: Add arguments element-wise.

3. np.all: Test whether all array elements along a given axis evaluate to True.

4. np.amax: Return the maximum of an array or maximum along an axis.

5. np.amin: Return the minimum of an array or minimum along an axis.

6. np.angle: Return the angle of the complex argument. #Z is a complex number or sequence of complex numbers

7. np.append: Append values to the end of an array.

8. np.argpartition: Perform an indirect partition along the given axis using the algorithm specified by the `kind` keyword.

9. np.around: Evenly round to the given number of decimals.

10. np.array2string: Return a string representation of an array.

11. np.ceil: Return the ceiling of the input, element-wise. #The ceil of the scalar `x` is the smallest integer `i`, such that
`i >= x`.

12. np.copy: Return an array copy of the given object.

13. np.concatenate: Join a sequence of arrays along an existing axis.

14. np.correlate: Cross-correlation of two 1-dimensional sequences. This function computes the correlation as generally defined.

15. np.datetime_as_string: Convert an array of datetimes into an array of strings.

16. np.delete: Return a new array with sub-arrays along an axis deleted.

17. np.diff: Calculate the n-th discrete difference along the given axis.

18. np.divide: Returns a true division of the inputs, element-wise.

19. np.extract: Return the elements of an array that satisfy some condition.

20. np.fmin: Element-wise minimum of array elements.
