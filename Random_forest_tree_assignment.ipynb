```python
#1 Iris dataset
```

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
iris
```

```python
dir(iris)
```

```python
print(iris.data.shape)
```

```python
df = pd.DataFrame(iris.data)
df.head()
```

```python
df["target"] = iris.target
```

```python
df[0:10]
```

```python
#Training, Modelling and prediction
```

```python
X = df.drop("target", axis = "columns")
y = df.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier (n_estimators = 20)
model.fit(X_train, y_train)
```

```python
X_test
```

```python
y_test
```

```python
model.score(X_test, y_test)
```

```python
y_predicted_value = model.predict(X_test)
```

```python
y_predicted_value 
```

```python
#confusion matrix
```

```python
from sklearn.metrics import confusion_matrix
con_max = confusion_matrix(y_test, y_predicted_value)
con_max
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize = (8,4))
sns.heatmap(con_max, annot = True)
plt.xlabel("Predicted_Value")
plt.ylabel("True_value")
```

```python
#2 diabetes datset
```

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
diabetes
```

```python
dir(diabetes)
```

```python
print(diabetes.data.shape)
```

```python
df_diabetes = pd.DataFrame(diabetes.data)
df_diabetes.head()
```

```python
df_diabetes["target"] = diabetes.target
```

```python
df_diabetes[0:10]
```

```python
#Training, Modelling and prediction
```

```python
X = df_diabetes.drop("target", axis = "columns")
y = df_diabetes.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier (n_estimators = 20)
model.fit(X_train, y_train)
```

```python
X_test
```

```python
y_test
```

```python
model.score(X_test, y_test)
```

```python
y_predicted_value = model.predict(X_test)
```

```python
y_predicted_value 
```

```python
#confusion matrix
```

```python
from sklearn.metrics import confusion_matrix
con_max = confusion_matrix(y_test, y_predicted_value)
con_max
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize = (12,9))
sns.heatmap(con_max, annot = True)
plt.xlabel("Predicted_Value")
plt.ylabel("True_value")
```
