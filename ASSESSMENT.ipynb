```python
#copy()
arr = np.array([1,2,3,4,6])
x = arr.copy()
arr[0] = 45
print(arr)
print(x)
```

```python
#view()
y= arr.view()
arr[1]= 26
print(arr)
print(y)
```

```python
arr = np.array([3,5,7,9,11])

x = arr.copy()
y = arr.view()

print(x.base)
print(y.base)               #base() to check if the copy owns the data
```

```python
#concatenate
arr1 = np.array([10,30,50])
arr2 = ([6,5,4])

arr = np.concatenate((arr1, arr2))
print(arr)
```

```python
#sort
arr = np.array([2,8,5,100])
print(np.sort(arr))
```

```python
#array_split()
data = np.array([1,3,5,7,9,11])
newdata = np.array_split(data, 3)
print(newdata)
```

```python
#where()
arr = np.array([1,2,3,4,5,4,4])
x = np.where(arr ==4)
print(x)
```

```python
#searchsorted()
arr = np.array([6,7,8,9])
x = np.searchsorted(arr, 7)
print(x)
```

```python
#choice()
from numpy import random
x = random.choice([3,5,7,9], p=[0.1, 0.3, 0.6, 0.0], size =(3,5))
print(x)
```

```python
#astype()
arr = np.array([1.1, 2.1, 3.1])
newarr =arr.astype('i')
print(newarr)
print(newarr.dtype)
```
