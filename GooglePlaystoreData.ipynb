**2) Clean the data set.**

```python
#Cleaning the dataset

chars_to_remove = ['+', ',', '$']
cols_to_clean = ['Installs', 'Price']

for cols in cols_to_clean:
    for chars in chars_to_remove:
        df[cols] = df[cols].apply(lambda x: x.replace(chars, ''))
df.head()
```

**3) Convert Install and Price column to float data type.**

```python
df = df[df.Installs!= 'free']
df= df[df.Price != 'Everyone']
df
#df.dropna([subset='free', 'everyone', ], inplace=True)
```

```python
df['Installs'] = df['Installs'].astype(float)
df['Price'] = df['Price'].astype(float) 
```

**5) Explore the distribution of app ratings.**

```python
df.groupby('Category')['Rating'].mean().sort_values(ascending=False)
```

```python
plot2 = df.groupby('App')['Rating'].mean()
binsize = 1
bins = np.arange(0, plot2.max()+binsize, binsize)
plt.figure(figsize=[10, 5])
plot2.plot(kind = 'hist', bins = bins)
plt.title('Distribution of App Ratings')
plt.show()
```

Majority of the apps have ratings close to 5 which is the highest possible rating.

**7) Plot to visualize the distribution of paid apps across different categories.**

```python
sns.stripplot(x = 'Price', y = 'Category', data = df)
plt.xticks(rotation = 90)
plt.figure(figsize = (30,15));
```

```python
#popular_app_cats = df[df.Category.isin(popular.head(10))]
popular_app_cats = df[df.Category.isin(['GAME','FAMILY','PHOTOGRAPHY','MEDICAL','TOOLS','FINANCE','LIFESTYLE','BUSINESS'])]
```

```python
popular_app_cats
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(15,8)
ax = sns.stripplot(x = popular_app_cats['Price'], y= popular_app_cats['Category'],jitter=True,linewidth=1)
ax.set_title('App pricing Category')
```
