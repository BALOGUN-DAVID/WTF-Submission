```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
df = load_iris()
df
```

```python
df_iris = pd.DataFrame(df.data, columns = df.feature_names)
df_iris
```

```python
df_target = pd.DataFrame(df.target, columns = ['Target'])
df_target
```

```python
df.target_names
```

```python
df_target = df_target.replace(to_replace = [0,1,2], value = df.target_names)
df_target
```

```python
iris = pd.concat([df_iris, df_target], axis = 1)
iris
```

```python
iris.describe()
```

```python
iris.info()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x="Target", data=iris, palette="viridis")
plt.xlabel('Iris Plant')
plt.show()
```

```python
# to show the outliers 
sns.boxplot(x='sepal length (cm)', data=iris)
plt.xlabel('Sepal Length (cm)')
plt.show()
```

```python
sns.boxplot(x='sepal width (cm)', data=iris)
plt.xlabel('Sepal Width (cm)')
plt.show()
```

```python
sns.boxplot(x='petal length (cm)', data=iris)
plt.xlabel('Petal Length (cm)')
plt.show()
```

```python
sns.boxplot(x='petal width (cm)', data=iris)
plt.xlabel('Petal Width (cm)')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier (n_estimators = 20)
```

```python
X = iris.drop(['Target'], axis = 1)
y = iris.Target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, prediction))
```

```python
print(confusion_matrix(y_test, prediction))
```

```python
conf = confusion_matrix(y_test, prediction)
```

```python
plt.figure(figsize = (10,7))
sns.heatmap(conf, annot = True)
plt.xlabel("Predicted Value")
plt.ylabel("Truth")
```

# Diabetes

```python
from sklearn.datasets import load_diabetes
```

```python
df1 = load_diabetes()
df1
```

```python
df_diabetes = pd.DataFrame(df1.data, columns = df1.feature_names)
df_diabetes
```

```python
df1_target = pd.DataFrame(df1.target, columns =['Target'])
df1_target
```

```python
diabetes = pd.concat([df_diabetes, df1_target], axis = 1)
diabetes
```

```python
diabetes.info()
```

```python
diabetes.describe()
```

```python
from sklearn.ensemble import RandomForestRegressor
model1 = RandomForestRegressor(n_estimators = 40)
```

```python
X1 = diabetes.drop(['Target'], axis = 1)
y1 = diabetes.Target
```

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size = 0.3, random_state = 42)
```

```python
model1.fit(X1_train, y1_train)
```

```python
prediction1 = model1.predict(X1_test)
prediction1
```

```python
y1_test
```

```python
sklearn.metrics.mean_squared_error(y1_test, prediction1)
```
